### With scikit learn

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(['selling_price'],axis=1), df["selling_price"], test_size=0.2)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop="first")

ohe.fit(x_train[["fuel","owner"]])

x_train=ohe.transform(x_train[["fuel","owner"]]).toarray()
x_test=ohe.transform(x_test[["fuel","owner"]]).toarray()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop="first", sparse_output=True)

ohe.fit(x_train[["fuel","owner"]])

x_train=ohe.transform(x_train[["fuel","owner"]])
x_test=ohe.transform(x_test[["fuel","owner"]])
```
